```
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
#from tensorflow.keras.preprocessing import image
#from IPython import display
```

```
!wget "https://i.pinimg.com/originals/46/36/d3/4636d34f40f37a4bf056bd373e25afd4.jpg" -O AMOL_IRAN.jpg
```

```
img = cv.imread('./AMOL_IRAN.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```
img = np.array(img)
print(type(img))
```

```
a = img.shape[0]
b = img.shape[1]
```

```
out = np.copy(img)
```

```
composition = input('please type a color composition like: \n\n rgb , gbr , rbg , bgr , brg , grb . . . even rgg . . . \n\n  as you wish...\n\n you chose ')
c1 = composition[0]
c2 = composition[1]
c3 = composition[2]
cols = [c1, c2, c3]
```

```
for x in range(0, 2):
    if cols[x] == 'r':
        color = img[:, :, 0]
    elif cols[x] == 'g':
        color = img[:, :, 1]
    elif cols[x] == 'b':
        color = img[:, :, 2]
    out[:, :, x] = color
```

```
plt.imshow(out)
plt.show
#display.Image(pic, width=500, height=500)
```
